#### Import Turicreate

```python
import turicreate
from math import log
```

#### Import King County House Price Dataset

```python
sales = turicreate.SFrame('home_data.sframe/')
```

#### Additional features which are functions of the original variables added to the dataset

```python
sales['bedrooms_squared'] = sales['bedrooms'].apply(lambda x:x**2)
sales['bed_bath_rooms'] = sales['bathrooms']*sales['bedrooms']
sales['log_sqft_living'] = sales['sqft_living'].apply(lambda x:log(x))
sales['lat_plus_long'] = sales['lat'] + sales['long']
```

#### A snapshot of the final dataset we would be working with in building the models

```python
sales
```

#### splitting the dataset into train & test datasets

```python
train_data,test_data = sales.random_split(.8,seed=0)
```

#### Getting the means of the newly added datasets

```python
print (test_data['bedrooms_squared'].mean())
print (test_data['bed_bath_rooms'].mean())
print (test_data['log_sqft_living'].mean())
print (test_data['lat_plus_long'].mean())
```

#### Multiple regression models

```python
model1 = turicreate.linear_regression.create(train_data, target = 'price', features = ['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long'], validation_set = None, verbose = False)
model2 = turicreate.linear_regression.create(train_data, target = 'price', features = ['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long', 'bed_bath_rooms'], validation_set = None, verbose = False)
model3 = turicreate.linear_regression.create(train_data, target = 'price', features = ['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long', 'bed_bath_rooms', 'bedrooms_squared','log_sqft_living','lat_plus_long'], validation_set = None, verbose = False)
```

#### Function for getting the RSS values from the rmse values for the training dataset

```python
def get_rss_from_rmse (model, data):
    RMSE_Model = model.evaluate(data);
    RSS_Model = RMSE_Model['rmse']*RMSE_Model['rmse']*len(data);
    return RSS_Model
```

```python
RSS_train_data_Model1 = get_rss_from_rmse(model1, train_data)
RSS_train_data_Model2= get_rss_from_rmse(model2, train_data)
RSS_train_data_Model3 = get_rss_from_rmse(model3, train_data)
print (RSS_train_data_Model1)
print (RSS_train_data_Model2)
print (RSS_train_data_Model3)
```

#### Function for getting the RSS values from the rmse values for the testing dataset

```python
RSS_test_data_Model1 = get_rss_from_rmse(model1, test_data)
RSS_test_data_Model2 = get_rss_from_rmse(model2, test_data)
RSS_test_data_Model3 = get_rss_from_rmse(model3, test_data)
print (RSS_test_data_Model1)
print (RSS_test_data_Model2)
print (RSS_test_data_Model3)
```

#### New Function for getting the RSS values(from their definition) on the training dataset

```python
def get_rss(model,data,output):
    predicted = model.predict(data);
    residual = output - predicted;
    rss = sum(residual*residual);
    return rss
```

```python
RSS_bydef_model1 = get_rss(model1,train_data,train_data['price'])
RSS_bydef_model2 = get_rss(model2,train_data,train_data['price'])
RSS_bydef_model3 = get_rss(model3,train_data,train_data['price'])
print(RSS_bydef_model1)
print(RSS_bydef_model2)
print(RSS_bydef_model3)
```

#### New Function for getting the RSS values(from their definition) on the testing dataset

```python
RSS_bydef_model1 = get_rss(model1,test_data,test_data['price'])
RSS_bydef_model2 = get_rss(model2,test_data,test_data['price'])
RSS_bydef_model3 = get_rss(model3,test_data,test_data['price'])
print(RSS_bydef_model1)
print(RSS_bydef_model2)
print(RSS_bydef_model3)
```
